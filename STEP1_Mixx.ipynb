# Classification of small molecules based on IC50 values

```python
action = []
for i in df1.activity:
    if float(i) <=100:
        action.append("active")
    elif float(i)<=5000:
        action.append("intermediate")
    else:
        action.append("inactive")
```

```python
#remove the number with < or > symbols
```

```python
df1
```

```python
action
```

```python
action = pd.Series(action, name = 'bioactivity')
```

# Joining the dataframes

```python
df2 = pd.concat([df1, action],axis = 1)
```
